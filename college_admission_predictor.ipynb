```python
# College Admission Predictor - IBM Cloud Jupyter Notebook
# Author: Rizwana

import pandas as pd

# Simulated student data
data = {
    'Name': ['Student A', 'Student B', 'Student C', 'Student D'],
    'Score (%)': [92, 85, 78, 66],
    'Category': ['General', 'OBC', 'SC', 'ST'],
    'Stream': ['Science', 'Commerce', 'Science', 'Arts']
}

df = pd.DataFrame(data)

# Define simple eligibility logic
def check_admission_eligibility(score):
    if score >= 90:
        return "Highly Eligible ✅"
    elif score >= 75:
        return "Moderately Eligible ✅"
    else:
        return "Not Eligible ❌"

# Add prediction column
df['Admission Status'] = df['Score (%)'].apply(check_admission_eligibility)

# Show results
df
```

```python
# Interactive prediction based on user input
try:
    your_score = int(input("Enter your Score (%): "))
    status = check_admission_eligibility(your_score)
    print(f"Your Admission Status: {status}")
except ValueError:
    print("Please enter a valid integer score.")
```
